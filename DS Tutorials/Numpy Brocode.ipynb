### Built-in Functions

```python
zeros = np.zeros((2,3,5))
print(zeros)
```

```python
Ones = np.ones((2,3,5))
print(Ones)
```

```python
Full = np.full((2,3,5),3)
print(Full)
```

```python
eye = np.eye(3)
print(eye)
```

```python
array = np.linspace(0,10,3)
print(array)
```

```python
array = np.array([[4, 2, 3, 1, 5],
                  [6, 7, 10, 9, 8]])

print(f'sum : {np.sum(array)}')
print(np.mean(array))
print(np.std(array))
print(np.var(array))
print(np.min(array))
print(np.max(array))
print(np.argmin(array))
print(np.argmax(array))
```

```python
ages = np.array([[21,17,19,23,26],
                [39,50,35,31,32]])

teenagers = ages[ages<20]
print(teenagers)
```

```python
adults = np.where(ages<18,ages,0)
print(adults)
```
