### Data Importing and cleaning

```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
```

```python
cols = ['fLength','fWidth','fSize','fConc', 'fConc1','fAsym','fM3Long','fM3Trans','fAlpha','fDist','class']
df = pd.read_csv('Magic.data', names = cols )
df.head()
```

```python
# df['class'].unique()
df['class'] = (df['class'] == 'g').astype(int)
# df.head()
# df[df['class'] == 1]
```

```python
df.head()
```

### Plotting

```python
for labels in cols[:-1]:
    plt.hist(df[df['class'] == 1][labels],color = 'blue',label = 'gamma',alpha = 0.7,density = True)
    plt.hist(df[df['class'] == 0][labels],color = 'red',label = 'hedron',alpha = 0.7,density = True)
    plt.title(labels)
    plt.ylabel('Probability')
    plt.xlabel(labels)
    plt.legend()
    plt.show()
```
