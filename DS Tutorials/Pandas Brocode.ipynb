### Series

```python
data = [101,102,103,104,105]

Series = pd.Series(data,index=["A","B","C","D","E"])
print(Series)
```

```python
print(Series.loc["C"])
print(Series.iloc[4])
```

```python
print(Series[Series >= 103])
```

```python
Calories = {'MON': 1750 , 'TUE': 1700 , 'WED': 1800 , 'THUR' : 2500 , 'FRI': 1850}
print(pd.Series(Calories))
```

```python
series = pd.Series(Calories)
print(series.loc['MON'])
print(series.loc['TUE'])
print(series.loc['WED'])
print(series.loc['THUR'])
```
